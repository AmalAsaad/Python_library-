```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,2*np.pi ,num=100)
y=np.cos(x)
z=np.sin(x)
```

```python
plt.plot(x,y ,"r--",label="cosine", linewidth=4)
plt.plot(x,z,"bx" , label="sine")
plt.title("the first")
plt.xlabel("time")
plt.ylabel("angel")
plt.xlim(0,2*np.pi)
plt.legend(frameon=True ,loc=5)
```
